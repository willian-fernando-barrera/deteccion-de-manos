## Loop/ciclos "for"

```python
numeros = [1, 2, 3, 4, 5]

for numero in numeros:
    print(numero)
```

## Loop/ciclo "While"

```python
contador = 5

while contador > 0:
    print(contador)
    contador = contador - 1
```

## Pruebas condicionales con if


```python
lenguaje = "c++"

if lenguaje == "python":
    print("El mejor lenguaje del mundo")
```
